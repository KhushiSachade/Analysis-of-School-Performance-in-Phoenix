```python
## for transfprmation of dataset

import pandas as pd
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv("cleaned.csv")


L_E = LabelEncoder()


for C in df.columns:
    if df[C].dtype == 'object':
        df[C] = L_E.fit_transform(df[C])


df.to_csv("new_transformed.csv", index=False)
```

```python
#firstly importing transformed file
#importing the pandas library
import pandas as g10pd

#reading the csv file into a df
g10df = g10pd.read_csv('/content/new_transformed.csv')
print(g10df)
```

```python
# Creating a new column 'Performance' and initializing with 'Low Performing'
g10df['Performance'] = 'Low Performing'

#now based on conditions assign high performnace to the schools
g10df.loc[(g10df['Drop-out rate %'] < 5) & (g10df['English score % (proficient)'] > 15) & (g10df['Maths score% (proficient)'] > 15), 'Performance'] = 'High Performing'

#displaying the updated DataFrame
print(g10df)
```

```python
g10df.to_csv("new_labelled.csv", index=False)
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier

label_df = g10pd.read_csv('/content/new_labelled.csv')

#selecting target variables and features.
g10_X = label_df[['English score % (proficient)', 'Maths score% (proficient)', 'Drop-out rate %', 'Total incidents of violence']]
g10_y = label_df['Performance']

#now creating kfold with folds=5
g10_kfold = KFold(n_splits=5, shuffle=True, random_state=42)

#creating Random Forest classifier
g10_clf = RandomForestClassifier()

#performing KFold cross-validation
g10_cv_scores = cross_val_score(g10_clf, g10_X, g10_y, cv=g10_kfold)

#printing the scores
print("Cross-validation scores:", g10_cv_scores)
print("Mean accuracy:", g10_cv_scores.mean())
```

```python
from sklearn.tree import DecisionTreeClassifier

#for creating decision Tree classifier
g10_decision_tree_clf = DecisionTreeClassifier()

#now fitting KFold cross-validation with Decision Tree classifier
g10_cv_scores_decision_tree = cross_val_score(g10_decision_tree_clf, g10_X, g10_y, cv=g10_kfold)

#printing the cross-validation scores
print("Cross-validation scores (Decision Tree Classifier):", g10_cv_scores_decision_tree)
print("Mean accuracy (Decision Tree Classifier):", g10_cv_scores_decision_tree.mean())
```

```python
from sklearn.neighbors import KNeighborsClassifier

#now evaluating with the KNN classifier
g10_knn_clf = KNeighborsClassifier()

#now fitting KFold cross-validation with KNN classifier
g10_cv_scores_knn = cross_val_score(g10_knn_clf, g10_X, g10_y, cv=g10_kfold)

# Print the cross-validation scores
print("Cross-validation scores (KNN Classifier):", g10_cv_scores_knn)
print("Mean accuracy (KNN Classifier):", g10_cv_scores_knn.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier

#creating the Random Forest classifier for evaluation
clf = RandomForestClassifier()

#performing K-Fold Cross Validation
cv_scores = cross_val_score(clf, g10_X, g10_y, cv=g10_kfold)

#printing the cross-validation scores
print("Cross-validation scores (Random Forest Classifier):", cv_scores)
print("Mean accuracy (Random Forest Classifier):", cv_scores.mean())
```

```python
from sklearn.svm import SVC

#now for evaluation creating the SVM classifier
g10_svm_clf = SVC()

#performing K-Fold Cross Validation with SVM classifier
g10_cv_scores_svm = cross_val_score(g10_svm_clf, g10_X, g10_y, cv=g10_kfold)

#printing the cross-validation scores
print("Cross-validation scores (SVM Classifier):", g10_cv_scores_svm)
print("Mean accuracy (SVM Classifier):", g10_cv_scores_svm.mean())
```
